$$ Question 1 $$

In matrix multiplication between an (n, k) matrix and a (k, m) matrix,

Number of Multiplications: n * m * k 

Number of Additions: n * m * (k - 1)

Each element of each row of first matrix will be multiplied by each element of each coloumn of second matrix

$$ Question 2 $$

```python
import numpy as np
import pylab as plt
import time
```

```python
def multiply_matrix(m1, m2):
    result = []

    for i in range(len(m1)):
        row = []
        for j in range(len(m2[0])):
            val = 0
            for k in range(len(m2)):
                val += m1[i][k] * m2[k][j]
            row.append(val)
        result.append(row)

    return result

m1 = [[1, 2, 3, 8], [8, 3, 2, 1]]  # 2x4
m2 = [[4, 5, 6], [5, 7, 7], [2, 5, 7], [2, 5, 7]]  # 4x3

t1 = time.time()
result = multiply_matrix(m1, m2)

print(result)
print("Time taken:", time.time() - t1)

print(f'Using numpy, {np.dot(m1,m2)}')
```

$$ Question 4 $$

The function:

$$ z(x, y) = x^2y + y^3 \sin(x) $$

The partial derivatives:

$$ \frac{\partial z}{\partial x} = 2xy + y^3 \cos(x) $$

$$ \frac{\partial z}{\partial y} = x^2 + 3y^2 \sin(x) $$

Gradient:

$$ \nabla z(x,y) = \frac{\partial z}{\partial x} + \frac{\partial z}{\partial y} = 2xy + y^3 \cos(x) + x^2 + 3y^2 \sin(x) $$

For (x,y) = (5,3)

$$ \nabla z(5,3) = 77.353 $$

$$ Question 6 $$

```python
import sympy as sym

x,y = sym.symbols('x y')


z = x**2 * y + y**3 * sym.sin(x)

# Calculate the partial derivatives
dz_dx, dz_dy = sym.diff(z, x), sym.diff(z, y)

print(f'Gradient, {dz_dx} + {dz_dy}')
```

$$ Question 7 $$

```python
class Student:
    def __init__(self, student_name, marks):
        self.name = student_name
        self.marks = marks

    def __repr__(self):
        return f"Student(name='{self.name}', marks={self.marks})"


class Branch:
    def __init__(self, students):
        self.students = students

    def __repr__(self):
        return f"Branch(students={self.students})"


class Year:
    def __init__(self, student_name, branches):
        self.name = student_name
        self.branches = branches

    def __repr__(self):
        return f"Year(name='{self.student_name}', branches={self.branches})"
```

```python
# student objects
student1 = Student('student_name1', {'math': 85, 'physics': 78, 'bio': 92})
student2 = Student('student_name2', {'math': 92, 'physics': 88, 'bio': 95})
student3 = Student('student_name3', {'math': 78, 'physics': 80, 'bio': 85})

# branch objects
branch1 = Branch([student1])
branch2 = Branch([student2, student3])

# Year object
year1 = Year('year1', [branch1, branch2])

year1.branches[0].students.append(Student('student_name4', {'math': 76, 'physics': 88, 'bio': 68}))

# overall database as a list of Year objects
student_data2 = [year1]


for year in student_data2:
    print(year.name)
    for branch in year.branches:
        print(f"  {branch}")
        print('')
        for student in branch.students:
            print(f"    {student}")
```

```python
x_val = np.arange(0.5, 100, 0.5)

# Functions
y1 = x_val
y2 = x_val**2
y3 = x_val**3/100
y4 = np.sin(x_val)
y5 = np.sin(x_val)/x_val
y6 = np.log(x_val)
y7 = np.exp(x_val)

y_val=[y1,y2,y3,y4,y5,y6,y7]

# Number of plots
num_plots = len(y_val)

# Plotting
plt.figure(figsize=(8, 12))

for i in range(num_plots):
    plt.subplot(num_plots, 1, i+1)
    plt.plot(x_val, y_val[i], label=f'y{i+1}')
    plt.xlabel('x_val')
    plt.ylabel('y_val')
    plt.title(f'y{i+1}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

matrix_20x5 = np.random.uniform(1, 2, (20, 5))

# DataFrame
data_frame = pd.DataFrame(matrix_20x5, columns=['a','b','c','d','e'])

# Finding coloumn with highest standard deviation and row with lowest mean
highest_std_values = data_frame.std()
highest_std = highest_std_values.idxmax()

lowest_mean = data_frame.mean(axis=1).idxmin()

print(f'highest standard deviation = {highest_std} with std = {highest_std_values[highest_std]}')
print(f'lowest mean = {lowest_mean}')
```

$$ Question 11 $$

```python
# Using previous 20x5 matrix and dataframe

# Adding a new column for total scores, just summing up all the grades
data_frame['f'] = data_frame[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

# fun column 'g', tagging students as 'LT8' or 'GT8' based on their total scores
data_frame['g'] = np.where(data_frame['f'] < 8, 'LT8', 'GT8')

# Counting number of students got the 'LT8' cool label
lt8_count = (data_frame['g'] == 'LT8').sum()

# Check the excitement in the scores / the standard deviation!
std_lt8 = data_frame.loc[data_frame['g'] == 'LT8', 'f'].std()
std_gt8 = data_frame.loc[data_frame['g'] == 'GT8', 'f'].std()


print(data_frame)
print("\nLT8 students:", lt8_count)
print("'LT8' scores standard deviation:", std_lt8)
print("'GT8' scores standard deviation:", std_gt8)
```
